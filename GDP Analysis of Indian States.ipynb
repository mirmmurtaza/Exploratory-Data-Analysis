### Why is (Primary + Secondary + Tertiary) not equal to total GDP?
> *I believe the answer lies in the fact that the GSDP is calculated by taking into account the* ***Taxes on Products*** *(which is added) and* ***Subsidies on products*** *(which is subtracted) whereas the* ***(Primary + Secondary + Tertiary)*** *results only in* ***TOTAL GSVA at basic prices*** *which is free from Taxes and Subsidies on products and hence are different in value.*

### *My Recommendations...*
*General Recommendation:*     **DIVIDE AND CONQUER**
>- *It is better to divide the responsibilities to generate GDP among all the sub-sectors in a fairly proportionate manner rather than having a single or a selected few sub-sectors as the drivers of the GDP generation. With this approach, even if a sub-sector is experiencing a slowdown, then their will be other sub-sectors compensating for it.*

***For C1:***

>- *Try to reduce the dependency from* ***'Agriculture, forestry and fishing'*** *and promote sub-sectors that are either less dependent or cimpletely independent os weather conditions.*

***For C2:***
>- ***'Electricity, gas, water supply & other utility services'***, *again, needs a targetted approach to tranform it into a major revenue generator.*
>- *The top sub-sectors are contributing in a proportionate manner but a better measure of proportion and balance should be achieved and make all the sub-sectors as revenue generators rather than a few.*
>- *India is in a developing phase and that requires development or contruction of the infrastructure. Therefore,* ***'Construction'*** *is one sub-sector that should be made to increase it's revenue generation as it has the potential.*

***For C3:***
>-  *The generation of revenue seems rather skewed with* ***'Agriculture, forestry and fishing'*** *occupying the top slot.* ***'Manufacturing'*** *comes at second but is lagging far behind. Therefore, efforts should be made to suppress this bias and make the revenue generation a collective effort.*
>- ***'Agriculture, forestry and fishing'*** *occupies the top slot but that is not necessarily good for the GSDP as this sub-sector depends heavily on the weather and climatic conditions which introduces uncertainties. ### *My Hypothesis:*

>- *On expected lines, with the increase in Per Capita GSDP, the dropout rates decreases and decrease in Per Capita GSDP leads to increase in Dropout rates. A theory can be that in families that are financially unstable, children tend to leave studies for work to meet the financial needs of the family.*

>- ***Primary Percentage Contribution*** *has a positive correlation with the* ***Dropout*** *rates. One possible explanation can be that the Primary Sector comprises mostly of manual labour related to Agriculture, Forestry, Fishing etc. and hence do not require sophisticated qualifications. It was also observed that Primary sector was the second major contributor in the not-so-developed states like Uttar Pradesh, Bihar etc. which have low Per Capita GSDP. All these factors are thereby creating an environment for the children to join these labour tasks.*

>- ***Secondary Percentage Contribution*** *has a negative correlation with the* ***Dropout*** *rates. One hypothesis that can explain such a phenomenon is that Secondary sector comprises mainly of formal jobs in the field Manufacturing, Electricity etc. that require qualifications and technical skills and hence children cannot be employed in that. These jobs are also better paid thereby minimizing the need for children to leave studies.*

>- ***Tertiary Percentage Contribution*** *shows a very unusual behaviour with the* ***Dropout*** *rates*
  <br/>***1.*** *With respect to* ***Primary Dropout Rate*** ---> ***Positive Correlation***
  <br/>***2.*** *With respect to* ***Upper Primary Dropout Rate*** ---> ***Slight Negative Correlation***
  <br/>***3.*** *With respect to* ***Secondary Dropout Rate*** ---> ***Virtually No Correlation*** 
<br/><br/>*The reason can be that Tertiary sector seems to have a combination of both formal jobs (that require qualifications) and informal jobs (that require less or no qualifications).*

>- ***Population*** *has a negative correlation with the* ***Dropout*** *rates. One possible explanation can be that increase in population is a real process that increases with time. But as time furthers, countries, including India, also furthers towards more development, infrastructure as well as economic, thereby creating better and formal work oppurtunities for the people which in turn decreases the dropout rates.*
